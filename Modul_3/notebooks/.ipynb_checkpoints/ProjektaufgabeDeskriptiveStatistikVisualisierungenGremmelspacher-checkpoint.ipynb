```python
# Wie ist die Altersverteilung der Kunden der San Francisco Public Library?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/Library_Usage.csv", usecols=["Age Range"])
df_filtered = df[df["Age Range"] != "Null"]
counts = df_filtered["Age Range"].value_counts()
print(counts)
print()
plt.figure(figsize=(9, 10))
sns.barplot(x=counts.index, y=counts.values, hue=counts.index, palette="Blues_r", legend=False)
plt.xlabel("Age Range")
plt.ylabel("Anzahl")
plt.title("Verteilung aller Altersgruppen")
plt.xticks(rotation=45)
plt.show()

```

```python
# Wie viele Senioren und Kinder sind Kunden der San Francisco Public Library?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/Library_Usage.csv", usecols=["Age Range"])
df_filtered = df[df["Age Range"] != "Null"]
selected_ranges = ["0 to 9 years", "65 to 74 years", "75 years and over"]
counts = df_filtered[df_filtered["Age Range"].isin(selected_ranges)]["Age Range"].value_counts()
print(counts)
print()
plt.figure(figsize=(6, 8))
sns.barplot(x=counts.index, y=counts.values, hue=counts.index, palette="Blues_r", legend=False)
plt.xlabel("Age Range")
plt.ylabel("Anzahl")
plt.title("Verteilung der Altersgruppen Senioren und Kinder")
plt.xticks(rotation=45)
plt.show()

```

```python
# Altersverteilung der Nutzer, die per Mail informiert werden möchten

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/Library_Usage.csv", usecols=["Age Range", "Notice Preference Definition"])
df_filtered = df[(df["Notice Preference Definition"] == "Email") & (df["Age Range"] != "Null")]
age_counts = df_filtered["Age Range"].value_counts()
print(age_counts)
print()
df_plot = age_counts.reset_index()
df_plot.columns = ["Age Range", "Count"]
plt.figure(figsize=(10, 6))
ax = sns.barplot(x="Age Range", y="Count", data=df_plot, hue="Age Range", palette="Blues_r", legend=False)
ax = plt.gca()
plt.xlabel("Age Range")
plt.ylabel("Anzahl")
plt.title("Altersverteilung der Kunden, die E-Mail-Benachrichtigung wünschen")
plt.xticks(rotation=45)
plt.show()
```

```python
# Wie alt sind Nutzer durchschnittlich, die 1) per Email, 2) per Telefon und 3) per Post informiert werden möchten?

import pandas as pd
import numpy as np

df = pd.read_csv("../data/Library_Usage.csv", usecols=["Age Range", "Notice Preference Definition"])
df = df[df["Notice Preference Definition"] != "Null"]

# Funktion zur Umwandlung von Altersgruppen in numerische Werte
def extract_mean_age(age_range):
    if pd.isna(age_range):  # Falls der Wert NaN ist, None zurückgeben
        return None
    numbers = [int(s) for s in age_range.split() if s.isdigit()]  # Alle Zahlen extrahieren
    return np.mean(numbers) if numbers else 100  # Mittelwert berechnen, sonst 100 (für "100 years and over")

# Durchschnittswerte für alle Altersgruppen berechnen
df["Age Numeric"] = df["Age Range"].apply(extract_mean_age)

# Durchschnittsalter für jede Benachrichtigungsmethode berechnen
avg_age = df.groupby("Notice Preference Definition")["Age Numeric"].mean().round(1)

# Ausgabe auf dem Bildschirm
for method, age in avg_age.items():
    print(f"Das Durchschnittsalter von Kunden, die {method}-Benachrichtigungen wünschen, beträgt: {age} Jahre")
print()
```
